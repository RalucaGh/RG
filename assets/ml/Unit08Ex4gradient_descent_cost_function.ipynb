## Calculating cost with gradient descent and learning rate
- Change the iteration and learning rate vaules and see the impact on cost. 
- Low iteration values with high learning rate (i.e. big steps) may lead to miss the global minimum
- Goal is to reach minimum cost with minimum iteration

```python
# code credit:codebasics https://codebasics.io/coming-soon
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 100  # Number of iterations
    n = len(x)
    learning_rate = 0.08  # Learning rate

    # Lists to store values for plotting
    m_values = []
    b_values = []
    cost_values = []

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1 / n) * sum([val**2 for val in (y - y_predicted)])
        md = -(2 / n) * sum(x * (y - y_predicted))
        bd = -(2 / n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        # Store values for plotting
        m_values.append(m_curr)
        b_values.append(b_curr)
        cost_values.append(cost)

        print(f"m {m_curr}, b {b_curr}, cost {cost}, iteration {i}")

    # Plotting
    plt.figure(figsize=(15, 5))

    # Plot 1: Data points and final regression line
    plt.subplot(1, 3, 1)
    plt.scatter(x, y, color="blue", label="Data Points")
    plt.plot(x, m_curr * x + b_curr, color="red", label="Final Regression Line")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Data Points and Regression Line")
    plt.legend()

    # Plot 2: Cost function over iterations
    plt.subplot(1, 3, 2)
    plt.plot(range(iterations), cost_values, color="green")
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("Cost Function Over Iterations")

    # Plot 3: m and b values over iterations
    plt.subplot(1, 3, 3)
    plt.plot(range(iterations), m_values, color="orange", label="m (slope)")
    plt.plot(range(iterations), b_values, color="purple", label="b (intercept)")
    plt.xlabel("Iterations")
    plt.ylabel("Value")
    plt.title("m and b Over Iterations")
    plt.legend()

    plt.tight_layout()
    plt.show()

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Run gradient descent
gradient_descent(x, y)
```
